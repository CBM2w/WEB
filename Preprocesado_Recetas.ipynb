# **Request de la url de dbpedia y conversión de datos obtenidos**

```python
import requests

url = 'https://api.dbpedia-spotlight.org/en/annotate'
dic_ingrediente = {"Id":[], "Nombre":[],"dbpedia_Uri":[],"dbpedia_Type":[]}

x = requests.get(url, params = {"text": recetas_muestra.loc[3]["Nombre"], "confidence" : "0.5"}, 
                     headers = {"Accept": "application/json"})

print(x.text)
```

```python
for row in range(recetas_muestra.shape[0]):
    x = requests.get(url, params = {"text": recetas_muestra.loc[row]["Nombre"], "confidence" : "0.5"}, 
                     headers = {"Accept": "application/json"})
    x = x.json()
    if 'Resources' in x:
        for i in range(len(x['Resources'])):
            dic_ingrediente["Id"].append(recetas_muestra.loc[row]["Id"])
            dic_ingrediente["Nombre"].append(x['Resources'][i]['@surfaceForm'])
            dic_ingrediente["dbpedia_Uri"].append(x['Resources'][i]['@URI'])
            dic_ingrediente["dbpedia_Type"].append(x['Resources'][i]['@types'])    
```

```python
print(dic_ingrediente.get("Nombre"))
print()
print("Tamaño del diccionario:", len(dic_ingrediente["Nombre"]))
print()
print("Uris de primeros ingredientes:\n", dic_ingrediente.get("dbpedia_Uri")[0:3])
```

```python
ingredienteDF = pd.DataFrame(dic_ingrediente)
ingredienteDF.head()
```
